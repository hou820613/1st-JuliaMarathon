# Julia 基礎圖表繪製: Plots 套件

## 作業 023：繪製 3D 函數

練習繪製 3D 函數：$z = x^2 - y^2$，$x$ 範圍於 $-5 \sim 5$，$y$ 範圍於 $-5 \sim 5$ 之間。（請完成以下程式碼）

```julia
using Pkg
Pkg.add("Plots")
```

```julia
using Plots
```

```julia
n = 1000
x = range(-5,5, length=n)
y = range(-5,5, length=n)
```

```julia
f(x,y) = x^2 - y^2
```

```julia
p = plot(x, y, f, st=:surface)
```

```julia
@gif for i in range(0, stop=2π, length=100)
    p = plot(x, y, f, st=:surface)
    plot!(p, camera=(15*cos(i), 40))
end
```
