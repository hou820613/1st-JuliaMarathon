## Day 20 作業：將 missing 值替代為其他值

由於現實世界中的資料可能有缺漏，因此資料中就會有 missing 值存在，而在資料科學及機器學習中，缺漏值的處理關係到分析或預測的結果。

在今天的作業中，請將資料集中的年齡缺漏值，依性別分別替換為資料集中男性或女性的年齡平均值。

本次作業使用的資料集：Kaggle Titanic: Machine Learning from Disaster 的 train 資料集

競賽主頁：[https://www.kaggle.com/c/titanic/overview](https://www.kaggle.com/c/titanic/overview)

請自行下載資料集 **train.csv**：[https://www.kaggle.com/c/titanic/data](https://www.kaggle.com/c/titanic/data)

```julia
using DataFrames, CSV
```

```julia
df = CSV.read("train.csv")
```

```julia
size(df)
```

呼叫 `describe()` 函式時，加上 `:nmissing` 參數，顯示各 column 的 missing 值數目。

可以看到 Age 共有 177 個缺漏值。

```julia
describe(df, :nmissing)
```

計算男性及女性的平均年齡。

【提示】使用 `skipmissing()` 去掉缺漏值後，再計算平均值。

```julia
using Statistics
```

```julia
male_age_avg = # 請輸入程式碼
df_Male = filter(row->row.Sex in ["male"], df)
df_MaleAge = df_Male[:,[:PassengerId, :Sex, :Age]]
male_age_avg = mean(skipmissing(df_MaleAge[:,:Age]))
println("男性的平均年齡 : ", male_age_avg," 歲")
```

```julia
female_age_avg = # 請輸入程式碼
df_Female = filter(row->row.Sex in ["female"], df)
df_FemaleAge = df_Female[:,[:PassengerId, :Sex, :Age]]
female_age_avg = mean(skipmissing(df_FemaleAge[:,:Age]))
println("女性的平均年齡 : ", female_age_avg ," 歲")
```

將 Age column 中的 missing 值，依性別替換為上面計算出來的平均年齡值。

【提示】可產生新的 column 來存放新的結果，而不取代掉原先的 Age column。

```julia
# 請輸入程式碼
array_male_age_coal = coalesce.(df_MaleAge[:,:Age], male_age_avg)
insertcols!(df_MaleAge, 4, :Age_Fix =>array_male_age_coal, makeunique=true)
array_female_age_coal = coalesce.(df_FemaleAge[:,:Age], female_age_avg)
insertcols!(df_FemaleAge, 4, :Age_Fix =>array_female_age_coal, makeunique=true)
```

```julia
df_ans = copy(df_MaleAge)
append!(df_ans,df_FemaleAge)
sort!(df_ans, :PassengerId)
show(df_ans, allrows=true,allcols=true)
```
